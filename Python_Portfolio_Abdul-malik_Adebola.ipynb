```python
## Numerical and tables

import numpy as np
import pandas as pd

## data visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fraudtest_df = pd.read_csv("fraud test.csv")
```

```python
## Check the Data types of all columns imported

fraudtest_df.info()
```

```python
## Confirm the Top values of the CSV data imported.

fraudtest_df.head()
```

```python
## Check the statistical analysis of the data per column

fraudtest_df.describe()
```

```python
## Check for Blank Cells on all the columns

fraudtest_df.isna().sum()
```

```python
## Set a dataframe for Fraud and Legitimate data.

fraud_df = fraudtest_df[fraudtest_df['is_fraud'] == 1]
legit_df = fraudtest_df[fraudtest_df['is_fraud'] == 0]
```

```python
# Get the number of fraudulent and legitimate transactions

fraudtest_df = pd.DataFrame(fraudtest_df)

fraudulent = fraud_df.shape[0]

legitimate = legit_df.shape[0]

print("Number of fraudulent: ", fraudulent) 
print("Number of Legitimate: ", legitimate)
```

```python
# chart to represent fraudulent and legitimate transactions

labels = ['fraudulent', 'legitimate']
sizes = [fraudulent, legitimate]
colors = ['#ff9999','#66b3ff']

fig1, ax1 = plt.subplots()
ax1.pie(sizes, colors = colors, labels=labels, autopct='%1.1f%%', startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Distribution of Fraudulent vs. Legitimate Transactions')
plt.show()

print(f"Number of Fraudulent Transactions: {fraudulent}")
print(f"Number of Legitimate Transactions: {legitimate}")
```

```python
#  Merchants involved in fraudulent transactions?

fraudulent_merchants = fraud_df['merchant'].value_counts().head(10)

fraudulent_merchants
```

```python
# Gender among customers involved mostly in fraudulent activities?

fruad_gender = fraud_df['gender'].value_counts()
legit_gender = legit_df['gender'].value_counts()

print("Gender involved in fruad are: ", fruad_gender)
print("Gender legitimate are: ",legit_gender)
```

```python
# Professions of customers involved with fraudulent transactions?

fraudulent_professionals = fraud_df['job'].value_counts()

print(fraudulent_professionals)
```

```python
# Professions of customers not involved with fraudulent transactions?

legit_professionals = legit_df['job'].value_counts()

print(legit_professionals)
```

```python
# fraudulent_transactions_by_city

fraud_city = fraud_df['state'].value_counts()

print(fraud_city)
```

```python
## fraudulent_transactions_by_gender

fraud_category_gender = fraud_df.groupby(['category', 'gender']).size()

print(fraud_category_gender)
```

```python
# Specific transaction amounts that are more likely to be fraudulent?

# Plot the distribution of transaction amounts for fraudulent transactions
plt.figure(figsize=(12, 6))

# Plot for fraudulent transactions
sns.histplot(fraud_df['amt'], kde=True, color='r', label='Fraudulent Transactions')

plt.title('Distribution of Transaction Amounts for Fraudulent Transactions')
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
